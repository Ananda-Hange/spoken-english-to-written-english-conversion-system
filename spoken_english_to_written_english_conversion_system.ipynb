```python
#create dictionary for spacial words:
numbers = {
                        "zero": 0,
                        "one" : 1,
                        "two": 2,
                        "three": 3,
                        "four": 4,
                        "five": 5,
                        "six": 6,
                        "seven": 7,
                        "eight": 8,
                        "nine": 9,
                        "ten": 10,
                        "twenty": 20,
                        "thirty": 30,
                        "forty": 40,
                        "fifty": 50,
                        "sixty": 60,
                        "seventy": 70,
                        "eighty": 80,
                        "ninety": 90,
                        "hundred": 100
}

tupels = {
                        "single":1,
                         "double":2,
                         "triple":3,
                         "quadruple":4,
                         "quintuple":5,
                         "sextuple":6,
                         "septuple":7,
                         "octuple":8,
                         "nonuple":9,
                         "decuple":10
}

General =  {
                          "C M": "CM",
                          "P M": "PM",
                          "D M": "DM",
                          "A M": "AM"
}


#creating function which will give ',' and '.' at the front or back of word.
def check_front_last(word):
    front=""
    last=""
    if(len(word)>1):
            if word[-1]==',' or word[-1]=='.':
                last=word[-1]
                word=word[:-1]
            if word[0]==',' or word[0]=='.':
                front=word[0]
                word=word[1:]
    return front,word,last

def show_output(paragraph,ouptut_para):
    print("\n[OUT]:The input Spoken English Paragraph: \n\n \" "+ paragraph+"\"")
    print("\nConverted Written English Paragraph: \n\n \"" +ouptut_para+"\"")
    
    
#This function will convert special words with given words or sentence.
def convert(paragraph):
    output_para=''
    words_in_para = paragraph.split()
    
    word_count = len(words_in_para)
    i = 0
    
    while i < word_count:
        #if i+1!= word_count:
            current_front,current_word,current_last = check_front_last(words_in_para[i])
            next_front,next_word,next_last = check_front_last(words_in_para[i+1])
        
            #when 'two dolers' ----> '$2' 
            if(current_word.lower() in numbers.keys() and (next_word.lower() =='dollars' or next_word.lower() =='dollar') ):
                output_para = output_para+' '+next_front+'$'+str(numbers[current_word.lower()])+next_last
                i = i+2
        
            #when word is of form Triple A ---> AAA
            elif(current_word.lower() in tupels.keys() and len(next_word)==1):
                output_para = output_para+' '+current_front+str(tupels[current_word.lower()]*next_word)+next_last
                i = i+2
            
             #if word is of form P M or C M
            elif current_word+' '+next_word in General.keys():
                
                output_para = output_para+' '+ next_front + current_word+next_word+next_last
                i = i+2
           
             #if above all condition not met    
            else:
                output_para = output_para+' '+words_in_para[i]
                i = i+1
                         
    return output_para          
```

```python
paragraph = input("\n[IN]:Enter Your paragraph of spoken english:\n\t")

if not paragraph:
    raise ValueError("[Error]: You entered nothing.")
    
print(60*'*'+ 'OUTPUT' + 60*'*')
out_para = convert(paragraph)
show_output(paragraph,out_para)
```
